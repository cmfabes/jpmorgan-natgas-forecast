```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt

CSV_PATH = "Nat_Gas.csv"  

#  Load & normalize 
df_raw = pd.read_csv(CSV_PATH)

def pick_cols(frame):
    lows = [c.lower().strip() for c in frame.columns]
    try:
        d = next(i for i,c in enumerate(lows) if c in ("date","dates","month","period"))
        p = next(i for i,c in enumerate(lows) if c in ("price","prices","value","close","settle"))
    except StopIteration:
        raise ValueError("CSV must have one date-like and one price-like column (e.g., Dates, Prices).")
    return frame.rename(columns={frame.columns[d]:"Dates", frame.columns[p]:"Prices"})[["Dates","Prices"]].copy()

df = pick_cols(df_raw)
try:
    df["Dates"] = pd.to_datetime(df["Dates"], format="%m/%d/%y", exact=True)
except Exception:
    df["Dates"] = pd.to_datetime(df["Dates"], format="mixed")
df["Prices"] = pd.to_numeric(df["Prices"], errors="coerce")
df = df.dropna().sort_values("Dates").reset_index(drop=True)

# Snap to month-end and average duplicates
df["Dates"] = df["Dates"] + pd.offsets.MonthEnd(0)
df = df.groupby("Dates", as_index=False)["Prices"].mean().sort_values("Dates")
if len(df) < 18:
    raise ValueError("Need at least ~18 monthly points for a defensible seasonal profile.")

#  daily history (log interpolation) 
me_idx = df["Dates"]
log_m  = np.log(df["Prices"].clip(lower=1e-8))
daily_idx = pd.date_range(me_idx.min().to_period("M").to_timestamp(how="start"),
                          me_idx.max().to_period("M").to_timestamp(how="end"),
                          freq="D")
log_daily = pd.Series(log_m.values, index=me_idx).reindex(daily_idx).interpolate(method="time")
hist_daily = pd.DataFrame({"Date": daily_idx, "Price": np.exp(log_daily.values)})

# Trend (daily)
hist_daily["logP"] = np.log(hist_daily["Price"].clip(lower=1e-8))
win = 30 if len(hist_daily) >= 60 else max(7, len(hist_daily)//6)
hist_daily["trend"] = hist_daily["logP"].rolling(win, center=True).median().bfill().ffill()

#  Seasonality (estimated from month-ends only) 
df_me = df.copy()
df_me["logP"] = np.log(df_me["Prices"].clip(lower=1e-8))
win_me = min(12, max(3, len(df_me)//6))
df_me["trend"] = df_me["logP"].rolling(win_me, center=True).median().bfill().ffill()
df_me["resid"] = df_me["logP"] - df_me["trend"]
df_me["month"] = df_me["Dates"].dt.month

seasonal_log = df_me.groupby("month")["resid"].mean()
seasonal_log = seasonal_log - seasonal_log.mean()             
seasonal_mult = np.exp(seasonal_log)

# Attach to daily
hist_daily["month"] = hist_daily["Date"].dt.month
hist_daily["fitted"] = np.exp(hist_daily["trend"] + hist_daily["month"].map(seasonal_log))

# Seasonality summary report
_short_labels = [
    "Winter peak","Heating demand","Late-winter high","Shoulder starts (inject)",
    "Injection phase","Summer lull","Power-gen lift","Stable low",
    "Pre-winter lift","Transition","Pre-winter prep","Heating season"
]

seasonality_report = pd.DataFrame({
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Seasonal Multiplier (Ã—)": seasonal_mult.reindex(range(1,13)).round(3).values,
    "Note": _short_labels
})

amp_pct = (seasonal_mult.max() - seasonal_mult.min()) * 100
snr = seasonal_log.std() / (df_me["resid"].std() + 1e-12)
pval = None if "anova_table" not in locals() else float(anova_table.loc["C(month)","PR(>F)"])

print("\n=== SEASONALITY ANALYSIS (Concise) ===\n")
print(seasonality_report.to_string(index=False))
print(f"\nAmplitude: {amp_pct:.2f}% | SNR: {snr:.3f}" + ("" if pval is None else f" | ANOVA p={pval:.6f}"))

if amp_pct < 2 or (pval is not None and pval > 0.10):
    print("\nTakeaway: Seasonality is negligible; use trend-only.")
else:
    print("\nTakeaway: Seasonality is material; keep month multipliers in the forward curve.")




# 12-month daily forecast (damped trend + guardrails)
H = 365
N = min(180, len(hist_daily))
x = np.arange(N)
y = hist_daily["trend"].tail(N).to_numpy()

slope, intercept = np.polyfit(x, y, 1)
last_trend = y[-1]
damp = 0.98
k = np.arange(1, H+1)
trend_increments = slope * (damp ** k)
f_trend = last_trend + np.cumsum(trend_increments)

last_day = hist_daily["Date"].max()
future_days = pd.date_range(last_day + pd.Timedelta(days=1), periods=H, freq="D")
f_seas = np.array([seasonal_log.get(m, 0.0) for m in future_days.month])

mu, sig = hist_daily["logP"].mean(), hist_daily["logP"].std()
f_log = np.clip(f_trend + f_seas, mu - 3*sig, mu + 3*sig)
future_daily = pd.DataFrame({"Date": future_days, "Forward": np.exp(f_log)})

forward_daily = pd.concat([
    hist_daily[["Date","fitted"]].rename(columns={"fitted":"Forward"}),
    future_daily
], ignore_index=True).set_index("Date")

# -------------------- Predictor API --------------------
def predict_price(target_date):
    t = pd.to_datetime(target_date)
    if t in forward_daily.index:
        return float(forward_daily.loc[t, "Forward"])
    if t < forward_daily.index.min():
        return float(forward_daily.iloc[0, 0])
    if t > forward_daily.index.max():
        return float(forward_daily.iloc[-1, 0])
    left = forward_daily.index[forward_daily.index.get_indexer([t], method="pad")[0]]
    right = forward_daily.index[forward_daily.index.get_indexer([t], method="backfill")[0]]
    if left == right:
        return float(forward_daily.loc[left, "Forward"])
    w = (t - left) / (right - left)
    return float((1-w)*forward_daily.loc[left, "Forward"] + w*forward_daily.loc[right, "Forward"])

# Optional quick prompt (uncomment to use in the notebook)
# try:
#     s = input("Enter date (YYYY-MM-DD): ").strip()
#     if s:
#         print(f"Estimated price on {s}: ${predict_price(s):.4f}")
# except EOFError:
#     pass

# -------------------- Plots --------------------
plt.figure()
plt.plot(hist_daily["Date"], hist_daily["Price"], label="Historical (Daily, interpolated)")
plt.plot(forward_daily.index, forward_daily["Forward"], label="Model (Daily, incl. 12-mo forecast)")
plt.title("Natural Gas: Daily History and 1-Year Forecast")
plt.xlabel("Date"); plt.ylabel("Price"); plt.legend(); plt.tight_layout(); plt.show()

plt.figure()
months = np.arange(1,13)
plt.bar(months, seasonal_mult.reindex(range(1,13)).values)
plt.xticks(months)
plt.title("Seasonal Multipliers (Month-of-Year)")
plt.xlabel("Month"); plt.ylabel("Multiplier (Ã—)"); plt.tight_layout(); plt.show()

plt.figure()
plt.plot(future_daily["Date"], future_daily["Forward"])
plt.title("Forward Curve (Next 12 Months, Daily, Damped)")
plt.xlabel("Date"); plt.ylabel("Forward Price"); plt.tight_layout(); plt.show()

# Demo: a couple of test lookups
print({
    "price_on_last_hist_month_end": f"{forward_daily.iloc[len(hist_daily)-1,0]:.4f}",
    "price_in_30_days":             f"{predict_price(hist_daily['Date'].max() + pd.Timedelta(days=30)):.4f}",
    "price_on_last_forecast_day":   f"{predict_price(future_daily['Date'].max()):.4f}"
})
```
